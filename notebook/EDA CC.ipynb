```python
def hex_to_int_safe(s):
    if not isinstance(s, str):
        return s
    cleaned = s.replace('CUS_', '').replace('_', '').replace('0x', '').replace('0X', '')
    if cleaned == '':
        return np.nan
    try:
        return int(cleaned, 16)
    except ValueError:
        return np.nan

df_copy['Customer_ID'] = df_copy['Customer_ID'].apply(hex_to_int_safe)
```

```python
# Convert ID column from hex to int
df_copy['ID'] = df_copy['ID'].apply(hex_to_int_safe)
df_copy['Customer_ID'] = df_copy['Customer_ID'].apply(hex_to_int_safe)
```

```python
def impute_values(series):
# normalize empty/blank names to NaN
    df_copy[series] = df_copy[series].replace(r'^\s*$', np.nan, regex=True)
    # df_copy['column'].str.contains(r'^\d+_$')
    # df_copy[series] = df_copy[series].replace(r'^\s*\d+_$', np.nan, regex=True)

    # get the first non-null name per Customer_ID_int
    first_name = (df_copy
              .loc[df_copy[series].notna()]
              .groupby('Customer_ID', sort=False)[series]
              .first())
    
    return first_name

# fill missing names by mapping from Customer_ID_int
df_copy['Name'] = df_copy['Name'].fillna(df_copy['Customer_ID'].map(impute_values('Name')))
df_copy['Name'] = df_copy['Name'].apply(lambda v: str(v).lstrip().rstrip() if not pd.isna(v) else np.nan)
df_copy['Age'] = df_copy['Age'].fillna(df_copy['Customer_ID'].map(impute_values('Age')))
df_copy['Occupation'] = df_copy['Occupation'].replace('_______', np.nan, regex=True)
df_copy['Occupation'] = df_copy['Occupation'].fillna(df_copy['Customer_ID'].map(impute_values('Occupation')))
df_copy['SSN'] = df_copy['SSN'].replace('#F%$D@*&8', np.nan)
df_copy['SSN'] = df_copy['SSN'].fillna(df_copy['Customer_ID'].map(impute_values('SSN')))
df_copy['Annual_Income'] = df_copy['Annual_Income'].astype(object).apply(lambda v: str(v).strip().strip('_') if not pd.isna(v) else np.nan)
df_copy['Annual_Income'] = df_copy['Annual_Income'].fillna(df_copy['Customer_ID'].map(impute_values('Annual_Income')))
df_copy['Annual_Income'] = df_copy['Annual_Income'].apply(lambda v: float(v) if not pd.isna(v) else np.nan)
df_copy['Monthly_Inhand_Salary'] = df_copy['Monthly_Inhand_Salary'].fillna(df_copy['Customer_ID'].map(impute_values('Monthly_Inhand_Salary')))  
df_copy['Num_of_Delayed_Payment'] = df_copy['Num_of_Delayed_Payment'].apply(lambda v: str(v).strip().strip('_') if not pd.isna(v) else np.nan)
df_copy['Num_of_Delayed_Payment'] = df_copy['Num_of_Delayed_Payment'].apply(lambda v: int(v) if not pd.isna(v) else np.nan)
df_copy['Payment_Behaviour'] = df_copy['Payment_Behaviour'].replace('!@9#%8', np.nan)
df_copy['Changed_Credit_Limit'] = df_copy['Changed_Credit_Limit'].replace('_', np.nan)
df_copy['Changed_Credit_Limit'] = df_copy['Changed_Credit_Limit'].apply(lambda v: float(v) if not pd.isna(v) else np.nan)
df_copy['Outstanding_Debt'] = df_copy['Outstanding_Debt'].apply(lambda v: str(v).strip().strip('_') if not pd.isna(v) else np.nan)
df_copy['Outstanding_Debt'] = df_copy['Outstanding_Debt'].apply(lambda v: float(v) if not pd.isna(v) else np.nan)

```

```python
# clean trailing/leading underscores and safely parse age values
# df_copy['Age'] = df_copy['Age'].astype(object).apply(lambda v: str(v).strip().strip('_') if not pd.isna(v) else np.nan)

def safe_parse_age(val):
	"""Return integer age if val represents a reasonable age (1-99), else np.nan."""
	if pd.isna(val):
		return np.nan
	# already numeric (int/float stored as string after previous step), try convert
	try:
		# allow strings like "23", "+23", "-1" etc.
		s = str(val).strip()
		if s == '':
			return np.nan
		if s.lstrip('+-').isdigit():
			ival = int(s)
			return ival if 0 < ival < 100 else np.nan
		# fallback: attempt float->int conversion when possible
		f = float(s)
		ival = int(f)
		return ival if 0 < ival < 100 else np.nan
	except Exception:
		return np.nan

df_copy['Age'] = df_copy['Age'].apply(safe_parse_age)

# build mapping and impute missing ages from Customer_ID where possible
age_map = impute_values('Age')  # returns Series indexed by Customer_ID
df_copy['Age'] = df_copy['Age'].fillna(df_copy['Customer_ID'].map(age_map))
df_copy['Num_Bank_Accounts'] = df_copy['Num_Bank_Accounts'].apply(safe_parse_age)
accounts_map = impute_values('Num_Bank_Accounts')
df_copy['Num_Bank_Accounts'] = df_copy['Num_Bank_Accounts'].fillna(df_copy['Customer_ID'].map(accounts_map))
df_copy['Num_Credit_Card'] = df_copy['Num_Credit_Card'].apply(safe_parse_age)
cards_map = impute_values('Num_Credit_Card')
df_copy['Num_Credit_Card'] = df_copy['Num_Credit_Card'].fillna(df_copy['Customer_ID'].map(cards_map))
df_copy['Num_of_Loan'] = df_copy['Num_of_Loan'].apply(safe_parse_age)
loans_map = impute_values('Num_of_Loan')
df_copy['Num_of_Loan'] = df_copy['Num_of_Loan'].fillna(df_copy['Customer_ID'].map(loans_map))
df_copy['Interest_Rate'] = df_copy['Interest_Rate'].apply(safe_parse_age)
ir_map = impute_values('Interest_Rate')
df_copy['Interest_Rate'] = df_copy['Interest_Rate'].fillna(df_copy['Customer_ID'].map(ir_map))
```

```python
import pandas as pd
import numpy as np
import re

def parse_age(age_str):
    if pd.isna(age_str):
        return None, None
    match = re.match(r"(\d+)\s*Years\s*and\s*(\d+)\s*Months", age_str)
    if match:
        return int(match.group(1)), int(match.group(2))
    return None, None

def format_age(years, months):
    return f"{years} Years and {months} Months"

def increment_age(years, months):
    months += 1
    if months > 11:
        months = 0
        years += 1
    return years, months

def decrement_age(years, months):
    months -= 1
    if months < 0:
        months = 11
        years -= 1
    return years, months

def impute_credit_history_age(df, col):
    # df = df.sort_values(['Customer_ID']).reset_index(drop=True)
    for cust_id, group in df.groupby('Customer_ID'):
        # Forward imputation
        prev_years, prev_months = None, None
        for i in group.index:
            val = df.at[i, col]
            if pd.isna(val) or val.strip() == '':
                if prev_years is not None:
                    new_years, new_months = increment_age(prev_years, prev_months)
                    df.at[i, col] = format_age(new_years, new_months)
                    prev_years, prev_months = new_years, new_months
                else:
                    # Placeholder None, will impute backward later
                    df.at[i, col] = None
            else:
                prev_years, prev_months = parse_age(val)

        # Backward imputation if first rows were None
        for idx in group.index[::-1]:
            if df.at[idx, col] is None or pd.isna(df.at[idx, col]) or df.at[idx, col].strip() == '':
                # Look forward to next non-null value
                next_idx = idx + 1
                if next_idx in df.index and (df.at[next_idx, col] not in [None, np.nan, '']):
                    next_years, next_months = parse_age(df.at[next_idx, col])
                    new_years, new_months = decrement_age(next_years, next_months)
                    df.at[idx, col] = format_age(new_years, new_months)
            else:
                # Once a valid value found going backward, stop
                break
    return df

# Execute on your DataFrame
df_copy = impute_credit_history_age(df_copy,'Credit_History_Age')
```
